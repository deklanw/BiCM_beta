```python
def solver2(x0, fun, stop_fun, fun_jac=None, tol=1e-8, eps=1e-3, max_steps=100, method='newton', verbose=False,
           regularise=False, full_return=False, linsearch=True, return_steps=False):
    """Find roots of eq. f = 0, using newton, quasinewton or dianati.
    """
    tic_all = time.time()
    toc_init = 0
    tic = time.time()
    # algorithm
    beta = .5  # to compute alpha
    n_steps = 0
    x = x0  # initial point
    # norm = np.linalg.norm(fun(x), ord=np.inf)
    norm = np.linalg.norm(fun(x))
    diff = 1
    if full_return:
        norm_seq = [norm]
    if verbose:
        print('\nx0 = {}'.format(x))
        print('|f(x0)| = {}'.format(norm))
    toc_init = time.time() - tic
    toc_alfa = 0
    toc_update = 0
    toc_dx = 0
    toc_jacfun = 0
    tic_loop = time.time()
    while norm > tol and diff > tol and n_steps < max_steps:  # stopping condition
        x_old = x  # save previous iteration
        # f jacobian
        tic = time.time()
        if method == 'newton':
            H = fun_jac(x)  # original jacobian
            # check the hessian is positive definite
            # l, e = np.linalg.eigh(H)
#             l, e = np.linalg.eig(H)
#             ml = np.min(l)
            # if it's not positive definite -> regularise
            # if ml < eps:
            #     regularise = True
            # regularisation
            if regularise == True:
                B = hessian_regulariser_function(H, eps)
                l, e = np.linalg.eigh(B)
                new_ml = np.min(l)
            else:
                B = H.__array__()
        elif method == 'quasinewton':
            # quasinewton hessian approximation
            B = fun_jac(x)  # Jacobian diagonal
            if regularise == True:
                B = np.maximum(B, B * 0 + 1e-8)
        toc_jacfun += time.time() - tic
        # discending direction computation
        tic = time.time()
        if method == 'newton':
            # print(l)
            # print('here', B, fun(x))
            # print('max', np.amax(B))
            # print('min', np.amin(B))
            # print('sum', np.sum(B))
            dx = np.linalg.solve(B, - fun(x))
        elif method == 'quasinewton':
            dx = - fun(x) / B
        elif method == 'fixed-point':
            dx = fun(x) - x
        toc_dx += time.time() - tic
        # backtraking line search
        tic = time.time()
        if linsearch == True:
            alfa = 1
            i = 0
            while sufficient_decrease_condition(stop_fun(x), stop_fun(x + alfa * dx), alfa, fun(x), dx) == False \
                    and i < 50:
                alfa *= beta
                i += 1
        else:
            alfa = 1
        toc_alfa += time.time() - tic
        tic = time.time()
        # solution update
        # direction= dx@fun(x).T
        if method in ['newton', 'quasinewton']:
            x = x + alfa * dx
        if method in ['fixed-point']:
            # x = alfa*(x + dx)
            x = x + alfa * dx
        toc_update += time.time() - tic
        # stopping condition computation
        # norm = np.linalg.norm(fun(x), ord=np.inf)
        norm = np.linalg.norm(fun(x))
        diff = np.linalg.norm(x - x_old)
        if full_return:
            norm_seq.append(norm)
        # step update
        n_steps += 1

        if verbose == True:
            print('step {}'.format(n_steps))
            print('alpha = {}'.format(alfa))
            print('fun = {}'.format(fun(x)))
            print('dx = {}'.format(dx))
            print('x = {}'.format(x))
            print('|f(x)| = {}'.format(norm))
            # print('x_old = {}'.format(x_old))
            # print('fun_old = {}'.format(fun(x_old)))
            # print('H_old = {}'.format(fun_jac(x_old)))
            # if method in ['newton', 'quasinewton']:
            # print('B = {}'.format(B))
    # if norm < tol:
    #     print('norm:', norm)
    # if diff < tol:
    #     print('diff:', diff)
    toc_loop = time.time() - tic_loop
    toc_all = time.time() - tic_all
    if verbose == True:
        print('Number of steps for convergence = {}'.format(n_steps))
        print('toc_init = {}'.format(toc_init))
        print('toc_jacfun = {}'.format(toc_jacfun))
        print('toc_alfa = {}'.format(toc_alfa))
        print('toc_dx = {}'.format(toc_dx))
        print('toc_update = {}'.format(toc_update))
        print('toc_loop = {}'.format(toc_loop))
        print('toc_all = {}'.format(toc_all))

    if full_return:
        return (x, toc_all, n_steps, np.array(norm_seq))
    if return_steps:
        return x, n_steps
    else:
        return x
```

```python
def eqs_old(xx, d_rows, d_cols, multiplier_rows, multiplier_cols, nrows):
    out_res = np.zeros(len(xx))
    xixa = vec2mat(xx[:nrows], xx[nrows:])
    xixa /= 1 + xixa
    out_res[:nrows] = (xixa * multiplier_cols).sum(axis=1) - d_rows
    out_res[nrows:] = (xixa.T * multiplier_rows).sum(axis=1) - d_cols
    return out_res
```

```python
def eqs_new(xx, d_rows, d_cols, multiplier_rows, multiplier_cols, nrows, ncols, out_res):
    out_res -= out_res

    for i in range(nrows):
        for j in range(ncols):
            x_ij = xx[nrows + j] * xx[i]
            multiplier_ij = x_ij / (1 + x_ij)
            out_res[i] += multiplier_cols[j] * multiplier_ij
            out_res[j + nrows] += multiplier_rows[i] * multiplier_ij

    for i in range(nrows):
        out_res[i] -= d_rows[i]
    for j in range(ncols):
        out_res[j + nrows] -= d_cols[j]
```

```python
@jit(nopython=True)
def check_sol2(biad_mat, avg_bicm, return_error=False, in_place=False):
    """
        This function prints the rows sums differences between two matrices, that originally are the biadjacency matrix and its bicm average matrix.
        The intended use of this is to check if an average matrix is actually a solution for a bipartite configuration model.

        If return_error is set to True, it returns 1 if the sum of the differences is bigger than 1.

        If in_place is set to True, it checks and sums also the total error entry by entry.
        The intended use of this is to check if two solutions are the same solution.
    """
    error = 0
    if (avg_bicm < 0).sum() != 0:
        print('NEGATIVE ENTRIES IN THE AVERAGE MATRIX!')
        error = 1
    rows_error_vec = np.abs(np.sum(biad_mat, axis=0) - np.sum(avg_bicm, axis=0))
    err_rows = np.max(rows_error_vec)
    print('max rows error =', err_rows)
    cols_error_vec = np.abs(np.sum(biad_mat, axis=1) - np.sum(avg_bicm, axis=1))
    err_cols = np.max(cols_error_vec)
    print('max columns error =', err_cols)
    tot_err = np.sum(rows_error_vec) + np.sum(cols_error_vec)
    print('total error =', tot_err)
    if tot_err > 1:
        error = 1
        print('WARNING total error > 1')
        if tot_err > 10:
            print('total error > 10')
    if err_rows + err_cols > 1:
        print('max error > 1')
        error = 1
        if err_rows + err_cols > 10:
            print('max error > 10')
    if in_place:
        diff_mat = np.abs(biad_mat - avg_bicm)
        print('In-place total error:', np.sum(diff_mat))
        print('In-place max error:', np.max(diff_mat))
    if return_error:
        return error
    else:
        return
```

```python
links_num = []
shapes = []
avg_times = []
stddevs = []
r_links_num = []
r_shapes = []
tol = 10
errors = {'newton': [], 'fixed-point': [], 'quasinewton': []}
times = {'newton': [], 'fixed-point': [], 'quasinewton': []}
stddevs = {'newton': [], 'fixed-point': [], 'quasinewton': []}
steps = {'newton': [], 'fixed-point': [], 'quasinewton': []}


methods = ['newton', 'quasinewton', 'fixed-point']
max_steps = {'newton': 100, 'quasinewton': 100, 'fixed-point': 200}

print('#######     BASCOMPTE    #######')
for dataset in working_datasets:
    print('\t Dataset:', dataset)
    biad_mat = np.loadtxt(bascompte_path + dataset + ".csv", delimiter=",", dtype=np.ubyte)
    
    links_num.append(biad_mat.sum())
    row_deg_seq = biad_mat.sum(1)
    r_row_deg_seq = np.unique(row_deg_seq)
    r_rows_len = len(r_row_deg_seq)
    col_deg_seq = biad_mat.sum(0)
    r_col_deg_seq = np.unique(col_deg_seq)
    r_cols_len = len(r_col_deg_seq)
    shapes.append(biad_mat.shape)
    r_shapes.append((r_rows_len, r_cols_len))
    r_links_num.append(r_row_deg_seq.sum())
    for method in methods:
        if method != 'quasinewton':
            linsearches = [True, False]
        else:
            linsearches = [True]
        for linsearch in linsearches:
            try:
                graph = BipartiteGraph()
                graph.set_degree_sequences((row_deg_seq, col_deg_seq))
                graph.solve_bicm(linsearch=linsearch, method=method, max_steps=max_steps[method], print_error=False)
                errors[method].append(check_sol_light(graph.x[graph.nonfixed_rows], graph.y[graph.nonfixed_cols], \
                                                      row_deg_seq[graph.nonfixed_rows] - len(graph.fixed_cols), \
                                                      col_deg_seq[graph.nonfixed_cols] - len(graph.fixed_rows), return_error=True))
                time_iter = %timeit -o (solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method]))
                times[method].append(time_iter.average)
                stddevs[method].append(time_iter.stdev)
                _, steps_this = solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method], return_steps=True)
                steps[method].append(steps_this)
            except:
                print(method + ' linsearch ' + repr(linsearch) + ' in dataset ' + repr(dataset) + ' didn\'t converge')
                traceback.print_exc()
                errors[method].append(1)
                times[method].append(0)
                stddevs[method].append(0)
                steps[method].append(0)

print('#######     INTERNATIONAL TRADE NETWORK    #######')
for year in np.arange(1995, 2011):
    print('Year ' + repr(year))
    biad_mat = np.loadtxt(wtw_path + repr(year) + '/M-matrix/Mcp_' + repr(year) + '.dat', dtype=np.ubyte)
    links_num.append(biad_mat.sum())
    row_deg_seq = biad_mat.sum(1)
    r_row_deg_seq = np.unique(row_deg_seq)
    r_rows_len = len(r_row_deg_seq)
    col_deg_seq = biad_mat.sum(0)
    r_col_deg_seq = np.unique(col_deg_seq)
    r_cols_len = len(r_col_deg_seq)
    shapes.append(biad_mat.shape)
    r_shapes.append((r_rows_len, r_cols_len))
    r_links_num.append(r_row_deg_seq.sum())
    
    for method in methods:
        if method != 'quasinewton':
            linsearches = [True, False]
        else:
            linsearches = [True]
        for linsearch in linsearches:
            try:
                graph = BipartiteGraph()
                graph.set_degree_sequences((row_deg_seq, col_deg_seq))
                graph.solve_bicm(linsearch=linsearch, method=method, max_steps=max_steps[method], print_error=False)
                errors[method].append(check_sol_light(graph.x[graph.nonfixed_rows], graph.y[graph.nonfixed_cols], \
                                                      row_deg_seq[graph.nonfixed_rows] - len(graph.fixed_cols), \
                                                      col_deg_seq[graph.nonfixed_cols] - len(graph.fixed_rows), return_error=True))
                time_iter = %timeit -o (solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method]))
                times[method].append(time_iter.average)
                stddevs[method].append(time_iter.stdev)
                _, steps_this = solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method], return_steps=True)
                steps[method].append(steps_this)
            except:
                print(method + ' linsearch ' + repr(linsearch) + ' in year ' + repr(year) + ' didn\'t converge')
                traceback.print_exc()
                errors[method].append(1)
                times[method].append(0)
                stddevs[method].append(0)
                steps[method].append(0)

print('#######     COLOMBIAN REGIONS EXPORTS    #######')
for year in np.arange(1992, 2016):
    print('Year:', year)
    biad_mat = np.load(colombia_regions_path + repr(year) + '/biad_mat.npy').astype(np.ubyte)
    while (np.any(~biad_mat.any(axis=1)) or np.any(~biad_mat.any(axis=0))):
        biad_mat = np.delete(biad_mat, np.where(~biad_mat.any(axis=1))[0], 0)
        biad_mat = np.delete(biad_mat, np.where(~biad_mat.any(axis=0))[0], 1)
    links_num.append(biad_mat.sum())
    row_deg_seq = biad_mat.sum(1)
    r_row_deg_seq = np.unique(row_deg_seq)
    r_rows_len = len(r_row_deg_seq)
    col_deg_seq = biad_mat.sum(0)
    r_col_deg_seq = np.unique(col_deg_seq)
    r_cols_len = len(r_col_deg_seq)
    shapes.append(biad_mat.shape)
    r_shapes.append((r_rows_len, r_cols_len))
    r_links_num.append(r_row_deg_seq.sum())
    
    for method in methods:
        if method != 'quasinewton':
            linsearches = [True, False]
        else:
            linsearches = [True]
        for linsearch in linsearches:
            try:
                graph = BipartiteGraph()
                graph.set_degree_sequences((row_deg_seq, col_deg_seq))
                graph.solve_bicm(linsearch=linsearch, method=method, max_steps=max_steps[method], print_error=False)
                errors[method].append(check_sol_light(graph.x[graph.nonfixed_rows], graph.y[graph.nonfixed_cols], \
                                                      row_deg_seq[graph.nonfixed_rows] - len(graph.fixed_cols), \
                                                      col_deg_seq[graph.nonfixed_cols] - len(graph.fixed_rows), return_error=True))
                time_iter = %timeit -o (solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method]))
                times[method].append(time_iter.average)
                stddevs[method].append(time_iter.stdev)
                _, steps_this = solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method], return_steps=True)
                steps[method].append(steps_this)
            except:
                print(method + ' linsearch ' + repr(linsearch) + ' in year ' + repr(year) + ' didn\'t converge')
                traceback.print_exc()
                errors[method].append(1)
                times[method].append(0)
                stddevs[method].append(0)
                steps[method].append(0)

print('#######     SOUTH AMERICAN EXPORTS    #######')
for country in countries:
    for year in years[country]:
        print(country, year)
        binary_df = np.loadtxt(py_simulations + country + "/" + repr(year) + "/binary_df.csv", delimiter=',', skiprows=1, dtype=str)
        biad_mat = binary_df[:, 1:].astype(np.ubyte)
        while (np.any(~biad_mat.any(axis=1)) or np.any(~biad_mat.any(axis=0))):
            biad_mat = np.delete(biad_mat, np.where(~biad_mat.any(axis=1))[0], 0)
            biad_mat = np.delete(biad_mat, np.where(~biad_mat.any(axis=0))[0], 1)
        links_num.append(biad_mat.sum())
        row_deg_seq = biad_mat.sum(1)
        r_row_deg_seq = np.unique(row_deg_seq)
        r_rows_len = len(r_row_deg_seq)
        col_deg_seq = biad_mat.sum(0)
        r_col_deg_seq = np.unique(col_deg_seq)
        r_cols_len = len(r_col_deg_seq)
        shapes.append(biad_mat.shape)
        r_shapes.append((r_rows_len, r_cols_len))
        r_links_num.append(r_row_deg_seq.sum())

        for method in methods:
            if method != 'quasinewton':
                linsearches = [True, False]
            else:
                linsearches = [True]
            for linsearch in linsearches:
                try:
                    graph = BipartiteGraph()
                    graph.set_degree_sequences((row_deg_seq, col_deg_seq))
                    graph.solve_bicm(linsearch=linsearch, method=method, max_steps=max_steps[method], print_error=False)
                    errors[method].append(check_sol_light(graph.x[graph.nonfixed_rows], graph.y[graph.nonfixed_cols], \
                                                          row_deg_seq[graph.nonfixed_rows] - len(graph.fixed_cols), \
                                                          col_deg_seq[graph.nonfixed_cols] - len(graph.fixed_rows), return_error=True))
                    time_iter = %timeit -o (solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                    method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method]))
                    times[method].append(time_iter.average)
                    stddevs[method].append(time_iter.stdev)
                    _, steps_this = solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                    method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method], return_steps=True)
                    steps[method].append(steps_this)
                except:
                    print(method + ' linsearch ' + repr(linsearch) + ' in country' + country + ' in year ' + repr(year) + ' didn\'t converge')
                    traceback.print_exc()
                    errors[method].append(1)
                    times[method].append(0)
                    stddevs[method].append(0)
                    steps[method].append(0)

print('#######     TWITTER DICIOTTI    #######')
span = 5
direct_dtype = np.dtype([('user_id', np.int64), ('tweet_id', np.int64), ('time', 'datetime64[m]'), ('tweet/retweet', 'i2')])
ed_list = np.loadtxt(twitter_diciotti + 'retweet_directed_netw.csv', delimiter=',', dtype=direct_dtype)
for day in range(11 - span + 1):
    if day <  9 - span:
        day_strings = ['2019-01-' + str(i) for i in range(23 + day, 23 + span + day)]# + ['2019-02-' + str(i).zfill(2) for i in range(1, 3)]
    else:
        day_strings = ['2019-01-' + str(i) for i in range(23 + day, 32)] + ['2019-02-' + str(i).zfill(2) 
                                                                            for i in range(1, day  - (9 - span) + 1)]
    print(day_strings)
    
    day_ed_list = np.array([tweet for tweet in ed_list if np.isin(str(tweet['time'])[:10], day_strings)], dtype=direct_dtype)
    
    tweets_edli = day_ed_list[day_ed_list['tweet/retweet'] == 0]
    retweets_edli = day_ed_list[day_ed_list['tweet/retweet'] == 1]
    
    tweet_user_id = np.unique(tweets_edli['user_id'])
    retweet_user_id = np.unique(retweets_edli['user_id'])
    
    tweet_id = np.unique(tweets_edli['tweet_id'])
    retweet_id = np.unique(retweets_edli['tweet_id'])
    
    all_users = list(set(tweet_user_id) | set(retweet_user_id))
    all_tweets = list(set(tweet_id) | set(retweet_id))
    
    biad_mat = np.zeros((len(all_users), len(all_tweets)), dtype=np.ubyte)
    
    user_pos_dict = {}
    us_count = 0
    for user in all_users:
        user_pos_dict.update({user: us_count})
        us_count += 1
        
    tweet_pos_dict = {}
    tw_count = 0
    for tweet in all_tweets:
        tweet_pos_dict.update({tweet: tw_count})
        tw_count += 1
    
    for tweet in tweets_edli:
        biad_mat[user_pos_dict[tweet['user_id']], tweet_pos_dict[tweet['tweet_id']]] = 1
    
    for retweet in retweets_edli:
        biad_mat[user_pos_dict[retweet['user_id']], tweet_pos_dict[retweet['tweet_id']]] = 1
    
    while (np.any(~biad_mat.any(axis=1)) or np.any(~biad_mat.any(axis=0))):
        biad_mat = np.delete(biad_mat, np.where(~biad_mat.any(axis=1))[0], 0)
        biad_mat = np.delete(biad_mat, np.where(~biad_mat.any(axis=0))[0], 1)
    
    links_num.append(biad_mat.sum())
    row_deg_seq = biad_mat.sum(1)
    r_row_deg_seq = np.unique(row_deg_seq)
    r_rows_len = len(r_row_deg_seq)
    col_deg_seq = biad_mat.sum(0)
    r_col_deg_seq = np.unique(col_deg_seq)
    r_cols_len = len(r_col_deg_seq)
    shapes.append(biad_mat.shape)
    r_shapes.append((r_rows_len, r_cols_len))
    r_links_num.append(r_row_deg_seq.sum())
    
    for method in methods:
        if method != 'quasinewton':
            linsearches = [True, False]
        else:
            linsearches = [True]
        for linsearch in linsearches:
            try:
                graph = BipartiteGraph()
                graph.set_degree_sequences((row_deg_seq, col_deg_seq))
                graph.solve_bicm(linsearch=linsearch, method=method, max_steps=max_steps[method], print_error=False)
                errors[method].append(check_sol_light(graph.x[graph.nonfixed_rows], graph.y[graph.nonfixed_cols], \
                                                      row_deg_seq[graph.nonfixed_rows] - len(graph.fixed_cols), \
                                                      col_deg_seq[graph.nonfixed_cols] - len(graph.fixed_rows), return_error=True))
                time_iter = %timeit -o (solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method]))
                times[method].append(time_iter.average)
                stddevs[method].append(time_iter.stdev)
                _, steps_this = solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                                method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method], return_steps=True)
                steps[method].append(steps_this)
            except:
                print(method + ' linsearch ' + repr(linsearch) + ' in day ' + repr(23 + day) + ' didn\'t converge')
                traceback.print_exc()
                errors[method].append(1)
                times[method].append(0)
                stddevs[method].append(0)
                steps[method].append(0)

print('#######     MEGA TWITTER MIGRANTI    #######')
edge_list = np.loadtxt('migranti_user_tweet_edges.csv', delimiter=',', dtype=np.dtype([('user', np.uint64), ('tweet', np.uint64), ('direction', 'i2')]))
edge_list_in = edge_list[edge_list['direction'] == 1]
users = np.unique(edge_list_in['user'])
tweets = np.unique(edge_list_in['tweet'])

biad_mat = np.zeros((len(users), len(tweets)), dtype='i1')

user_pos_dict = {}
us_count = 0
for user in users:
    user_pos_dict.update({user: us_count})
    us_count += 1

tweet_pos_dict = {}
tw_count = 0
for tweet in tweets:
    tweet_pos_dict.update({tweet: tw_count})
    tw_count += 1

for tweet in edge_list_in:
    biad_mat[user_pos_dict[tweet['user']], tweet_pos_dict[tweet['tweet']]] = 1

while (np.any(~biad_mat.any(axis=1)) or np.any(~biad_mat.any(axis=0))):
    biad_mat = np.delete(biad_mat, np.where(~biad_mat.any(axis=1))[0], 0)
    biad_mat = np.delete(biad_mat, np.where(~biad_mat.any(axis=0))[0], 1)

links_num.append(biad_mat.sum())
row_deg_seq = biad_mat.sum(1)
r_row_deg_seq = np.unique(row_deg_seq)
r_rows_len = len(r_row_deg_seq)
col_deg_seq = biad_mat.sum(0)
r_col_deg_seq = np.unique(col_deg_seq)
r_cols_len = len(r_col_deg_seq)
shapes.append(biad_mat.shape)
r_shapes.append((r_rows_len, r_cols_len))
r_links_num.append(r_row_deg_seq.sum())

for method in methods:
    if method != 'quasinewton':
        linsearches = [True, False]
    else:
        linsearches = [True]
    for linsearch in linsearches:
        try:
            graph = BipartiteGraph()
            graph.set_degree_sequences((row_deg_seq, col_deg_seq))
            graph.solve_bicm(linsearch=linsearch, method=method, max_steps=max_steps[method], print_error=False)
            errors[method].append(check_sol_light(graph.x[graph.nonfixed_rows], graph.y[graph.nonfixed_cols], \
                                                  row_deg_seq[graph.nonfixed_rows] - len(graph.fixed_cols), \
                                                  col_deg_seq[graph.nonfixed_cols] - len(graph.fixed_rows), return_error=True))
            time_iter = %timeit -o (solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                            method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method]))
            times[method].append(time_iter.average)
            stddevs[method].append(time_iter.stdev)
            _, steps_this = solver2(graph.x0, graph.fun, graph.stop_fun, fun_jac=graph.fun_jac,\
                                            method=method, verbose=False, linsearch=linsearch, max_steps=max_steps[method], return_steps=True)
            steps[method].append(steps_this)
        except:
            print(method + ' linsearch ' + repr(linsearch) + ' in mega twitter didn\'t converge')
            traceback.print_exc()
            errors[method].append(1)
            times[method].append(0)
            stddevs[method].append(0)
            steps[method].append(0)
```

```python
links_num = []
shapes = []
avg_times = []
stddevs = []
r_links_num = []
r_shapes = []

for tol in range(6, 15):
    print('Tolerance:', tol)
    
    iter_errors = []
    iter_times = []
    iter_stddevs = []
    
    newton_errors = []
    newton_times = []
    newton_stddevs = []
    for dataset in working_datasets:
        print('\t Dataset:', dataset)
        biad_mat = np.loadtxt(bascompte_path + dataset + ".csv", delimiter=",", dtype=np.ubyte)
        
        if tol == 6:
            links_num.append(biad_mat.sum())
            row_deg_seq = biad_mat.sum(1)
            r_row_deg_seq = np.unique(row_deg_seq)
            r_rows_len = len(r_row_deg_seq)
            col_deg_seq = biad_mat.sum(0)
            r_col_deg_seq = np.unique(col_deg_seq)
            r_cols_len = len(r_col_deg_seq)
#             print(links_num[-1])
            shapes.append(biad_mat.shape)
#             print(shapes[-1])
            r_shapes.append((r_rows_len, r_cols_len))
#             print(r_shapes[-1])
            r_links_num.append(r_row_deg_seq.sum())

        try:
            time_iter = %timeit -o (bicm_calculator(biad_mat, method='iterative', tolerance=10 ** - tol))
            avg_mat = bicm_calculator(biad_mat, method='iterative', tolerance=10 ** - tol)
            iter_errors.append(check_sol(avg_mat, biad_mat, return_error=True))
            iter_times.append(time_iter.average)
            iter_stddevs.append(time_iter.stdev)
        except:
            print('Iterative in dataset ' + repr(dataset) + ' didn\'t converge')
            iter_errors.append(1)
            iter_times.append(0)
            iter_stddevs.append(0)
        
        try:
            time_newton = %timeit -o (bicm_calculator(biad_mat, method='newton', tolerance=10 ** - tol))
            avg_mat = bicm_calculator(biad_mat, method='newton', tolerance=10 ** - tol)
            newton_errors.append(check_sol(avg_mat, biad_mat, return_error=True))
            newton_times.append(time_newton.average)
            newton_stddevs.append(time_newton.stdev)
        except:
            print('Newton in dataset ' + repr(dataset) + ' didn\'t converge')
            newton_errors.append(1)
            newton_times.append(0)
            newton_stddevs.append(0)
    
    if tol == 6:
        widths = np.array([min(shape) for shape in r_shapes])
        lengths = np.array([max(shape) for shape in r_shapes])
    
    ok_newton_loc = np.where(np.array(newton_errors) == 0)[0]
    err_newton_loc = np.where(np.array(newton_errors) == 1)[0]
    ok_iter_loc = np.where(np.array(iter_errors) == 0)[0]
    err_iter_loc = np.where(np.array(iter_errors) == 1)[0]
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(24, 24))
    axes[0, 0].scatter(np.log10(links_num)[ok_iter_loc], np.log10(iter_times)[ok_iter_loc], c=iter_color, label='iter')
    axes[0, 0].scatter(np.log10(links_num)[err_iter_loc], np.ones(len(err_iter_loc)) * np.average(np.log10(iter_times)[ok_iter_loc]), 
                       marker='x', c=iter_color, label='failed iter')
    axes[0, 0].scatter(np.log10(links_num)[ok_newton_loc], np.log10(newton_times)[ok_newton_loc], c=newton_color, label='newton')
    axes[0, 0].scatter(np.log10(links_num)[err_newton_loc], np.ones(len(err_newton_loc)) * np.average(np.log10(newton_times)[ok_newton_loc]), 
                       marker='x', c=newton_color, label='failed newton')
    axes[0, 0].set_xlabel('log10(Links number)')
    axes[0, 0].set_title('Tolerance: ' + repr(10 ** - tol))
    axes[0, 1].scatter(np.log10(r_links_num)[ok_iter_loc], np.log10(iter_times)[ok_iter_loc], c=iter_color, label='iter')
    axes[0, 1].scatter(np.log10(r_links_num)[err_iter_loc], np.ones(len(err_iter_loc)) * np.average(np.log10(iter_times)[ok_iter_loc]), 
                       marker='x', c=iter_color, label='failed iter')
    axes[0, 1].scatter(np.log10(r_links_num)[ok_newton_loc], np.log10(newton_times)[ok_newton_loc], c=newton_color, label='newton')
    axes[0, 1].scatter(np.log10(r_links_num)[err_newton_loc], np.ones(len(err_newton_loc)) * np.average(np.log10(newton_times)[ok_newton_loc]), 
                       marker='x', c=newton_color, label='failed newton')
    axes[0, 1].set_xlabel('log10(Links number after reduction)')
    axes[0, 1].set_title('Tolerance: ' + repr(10 ** - tol))
    axes[1, 0].scatter(np.log10(lengths)[ok_iter_loc], np.log10(iter_times)[ok_iter_loc], c=iter_color, label='iter')
    axes[1, 0].scatter(np.log10(lengths)[err_iter_loc], np.ones(len(err_iter_loc)) * np.average(np.log10(iter_times)[ok_iter_loc]), 
                       marker='x', c=iter_color, label='failed iter')
    axes[1, 0].scatter(np.log10(lengths)[ok_newton_loc], np.log10(newton_times)[ok_newton_loc], c=newton_color, label='newton')
    axes[1, 0].scatter(np.log10(lengths)[err_newton_loc], np.ones(len(err_newton_loc)) * np.average(np.log10(newton_times)[ok_newton_loc]), 
                       marker='x', c=newton_color, label='failed newton')
    axes[1, 0].set_xlabel('log10(Longer layer length (reduced))')
    axes[1, 0].set_title('Tolerance: ' + repr(10 ** - tol))
    axes[1, 1].scatter(np.log10(widths)[ok_iter_loc], np.log10(iter_times)[ok_iter_loc], c=iter_color, label='iter')
    axes[1, 1].scatter(np.log10(widths)[err_iter_loc], np.ones(len(err_iter_loc)) * np.average(np.log10(iter_times)[ok_iter_loc]), 
                       marker='x', c=iter_color, label='failed iter')
    axes[1, 1].scatter(np.log10(widths)[ok_newton_loc], np.log10(newton_times)[ok_newton_loc], c=newton_color, label='newton')
    axes[1, 1].scatter(np.log10(widths)[err_newton_loc], np.ones(len(err_newton_loc)) * np.average(np.log10(newton_times)[ok_newton_loc]), 
                       marker='x', c=newton_color, label='failed newton')
    axes[1, 1].set_xlabel('log10(Shorter layer length (reduced))')
    axes[1, 1].set_title('Tolerance: ' + repr(10 ** - tol))
    for i in range(2):
        for j in range(2):
            axes[i, j].legend(loc=0)
            axes[i, j].set_ylabel('log10(Time elapsed)')
    plt.show()
```
