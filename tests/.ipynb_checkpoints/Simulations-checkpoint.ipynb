```python
def solver2(x0, fun, stop_fun, fun_jac=None, tol=1e-8, eps=1e-3, max_steps=100, method='newton', verbose=False,
           regularise=False, full_return=False, linsearch=True, return_steps=False):
    """Find roots of eq. f = 0, using newton, quasinewton or dianati.
    """
    tic_all = time.time()
    toc_init = 0
    tic = time.time()
    # algorithm
    beta = .5  # to compute alpha
    n_steps = 0
    x = x0  # initial point
    # norm = np.linalg.norm(fun(x), ord=np.inf)
    norm = np.linalg.norm(fun(x))
    diff = 1
    if full_return:
        norm_seq = [norm]
    if verbose:
        print('\nx0 = {}'.format(x))
        print('|f(x0)| = {}'.format(norm))
    toc_init = time.time() - tic
    toc_alfa = 0
    toc_update = 0
    toc_dx = 0
    toc_jacfun = 0
    tic_loop = time.time()
    while norm > tol and diff > tol and n_steps < max_steps:  # stopping condition
        x_old = x  # save previous iteration
        # f jacobian
        tic = time.time()
        if method == 'newton':
            H = fun_jac(x)  # original jacobian
            # check the hessian is positive definite
            # l, e = np.linalg.eigh(H)
#             l, e = np.linalg.eig(H)
#             ml = np.min(l)
            # if it's not positive definite -> regularise
            # if ml < eps:
            #     regularise = True
            # regularisation
            if regularise == True:
                B = hessian_regulariser_function(H, eps)
                l, e = np.linalg.eigh(B)
                new_ml = np.min(l)
            else:
                B = H.__array__()
        elif method == 'quasinewton':
            # quasinewton hessian approximation
            B = fun_jac(x)  # Jacobian diagonal
            if regularise == True:
                B = np.maximum(B, B * 0 + 1e-8)
        toc_jacfun += time.time() - tic
        # discending direction computation
        tic = time.time()
        if method == 'newton':
            # print(l)
            # print('here', B, fun(x))
            # print('max', np.amax(B))
            # print('min', np.amin(B))
            # print('sum', np.sum(B))
            dx = np.linalg.solve(B, - fun(x))
        elif method == 'quasinewton':
            dx = - fun(x) / B
        elif method == 'fixed-point':
            dx = fun(x) - x
        toc_dx += time.time() - tic
        # backtraking line search
        tic = time.time()
        if linsearch == True:
            alfa = 1
            i = 0
            while sufficient_decrease_condition(stop_fun(x), stop_fun(x + alfa * dx), alfa, fun(x), dx) == False \
                    and i < 50:
                alfa *= beta
                i += 1
        else:
            alfa = 1
        toc_alfa += time.time() - tic
        tic = time.time()
        # solution update
        # direction= dx@fun(x).T
        if method in ['newton', 'quasinewton']:
            x = x + alfa * dx
        if method in ['fixed-point']:
            # x = alfa*(x + dx)
            x = x + alfa * dx
        toc_update += time.time() - tic
        # stopping condition computation
        # norm = np.linalg.norm(fun(x), ord=np.inf)
        norm = np.linalg.norm(fun(x))
        diff = np.linalg.norm(x - x_old)
        if full_return:
            norm_seq.append(norm)
        # step update
        n_steps += 1

        if verbose == True:
            print('step {}'.format(n_steps))
            print('alpha = {}'.format(alfa))
            print('fun = {}'.format(fun(x)))
            print('dx = {}'.format(dx))
            print('x = {}'.format(x))
            print('|f(x)| = {}'.format(norm))
            # print('x_old = {}'.format(x_old))
            # print('fun_old = {}'.format(fun(x_old)))
            # print('H_old = {}'.format(fun_jac(x_old)))
            # if method in ['newton', 'quasinewton']:
            # print('B = {}'.format(B))
    # if norm < tol:
    #     print('norm:', norm)
    # if diff < tol:
    #     print('diff:', diff)
    toc_loop = time.time() - tic_loop
    toc_all = time.time() - tic_all
    if verbose == True:
        print('Number of steps for convergence = {}'.format(n_steps))
        print('toc_init = {}'.format(toc_init))
        print('toc_jacfun = {}'.format(toc_jacfun))
        print('toc_alfa = {}'.format(toc_alfa))
        print('toc_dx = {}'.format(toc_dx))
        print('toc_update = {}'.format(toc_update))
        print('toc_loop = {}'.format(toc_loop))
        print('toc_all = {}'.format(toc_all))

    if full_return:
        return (x, toc_all, n_steps, np.array(norm_seq))
    if return_steps:
        return x, n_steps
    else:
        return x
```

```python
times_dict = {
    'newton linsearch': [],
    'newton no linsearch': [],
    'quasinewton linsearch': [],
    'fixed-point linsearch': [],
    'fixed-point no linsearch': []
}

stddevs_dict = {
    'newton linsearch': [],
    'newton no linsearch': [],
    'quasinewton linsearch': [],
    'fixed-point linsearch': [],
    'fixed-point no linsearch': []
}

errors_dict = {
    'newton linsearch': [],
    'newton no linsearch': [],
    'quasinewton linsearch': [],
    'fixed-point linsearch': [],
    'fixed-point no linsearch': []
}

steps_dict = {
    'newton linsearch': [],
    'newton no linsearch': [],
    'quasinewton linsearch': [],
    'fixed-point linsearch': [],
    'fixed-point no linsearch': []
}
```

```python
for method_i in range(len(methods)):
    if method_i < 1:
        for time_elapsed_i in range(len(times[methods[method_i]])):
            if time_elapsed_i % 2 == 0:
                times_dict['newton linsearch'].append(times[methods[method_i]][time_elapsed_i])
                stddevs_dict['newton linsearch'].append(stddevs[methods[method_i]][time_elapsed_i])
                errors_dict['newton linsearch'].append(errors[methods[method_i]][time_elapsed_i])
                steps_dict['newton linsearch'].append(steps[methods[method_i]][time_elapsed_i])
            else:
                times_dict['newton no linsearch'].append(times[methods[method_i]][time_elapsed_i])
                stddevs_dict['newton no linsearch'].append(stddevs[methods[method_i]][time_elapsed_i])
                errors_dict['newton no linsearch'].append(errors[methods[method_i]][time_elapsed_i])
                steps_dict['newton no linsearch'].append(steps[methods[method_i]][time_elapsed_i])
    elif method_i < 2:
        for time_elapsed_i in range(len(times[methods[method_i]])):
            times_dict['quasinewton linsearch'].append(times[methods[method_i]][time_elapsed_i])
            stddevs_dict['quasinewton linsearch'].append(stddevs[methods[method_i]][time_elapsed_i])
            errors_dict['quasinewton linsearch'].append(errors[methods[method_i]][time_elapsed_i])
            steps_dict['quasinewton linsearch'].append(steps[methods[method_i]][time_elapsed_i])
    else:
        for time_elapsed_i in range(len(times[methods[method_i]])):
            if time_elapsed_i % 2 == 0:
                times_dict['fixed-point linsearch'].append(times[methods[method_i]][time_elapsed_i])
                stddevs_dict['fixed-point linsearch'].append(stddevs[methods[method_i]][time_elapsed_i])
                errors_dict['fixed-point linsearch'].append(errors[methods[method_i]][time_elapsed_i])
                steps_dict['fixed-point linsearch'].append(steps[methods[method_i]][time_elapsed_i])
            else:
                times_dict['fixed-point no linsearch'].append(times[methods[method_i]][time_elapsed_i])
                stddevs_dict['fixed-point no linsearch'].append(stddevs[methods[method_i]][time_elapsed_i])
                errors_dict['fixed-point no linsearch'].append(errors[methods[method_i]][time_elapsed_i])
                steps_dict['fixed-point no linsearch'].append(steps[methods[method_i]][time_elapsed_i])
```
